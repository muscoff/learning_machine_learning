```python
import pandas as pd
```

```python
video_data = pd.read_csv('vgsales.csv')
```

```python
# video_data.describe()
```

```python
video_data.dropna(inplace=True)
```

```python
video_data.columns
```

```python
platform = video_data['Platform'].unique()
year = video_data['Year'].unique()
genre = video_data['Genre'].unique()
publisher = video_data['Publisher'].unique()
```

```python
platform_sales = []

for i in video_data.index:
    if len(platform_sales) == 0:
        obj = {
            'Platform': video_data.loc[i, 'Platform'],
            'NA_Sales': int(video_data.loc[i, 'NA_Sales']),
            'EU_Sales': int(video_data.loc[i, 'EU_Sales']),
            'JP_Sales': int(video_data.loc[i, 'JP_Sales']),
            'Global_Sales': int(video_data.loc[i, 'Global_Sales'])
        }
        platform_sales.append(obj)
    else:
        if video_data.loc[i, 'Platform'] in [obj['Platform'] for obj in platform_sales]:
            index = [obj['Platform'] for obj in platform_sales].index(video_data.loc[i, 'Platform'])
            platform_sales[index]['NA_Sales'] += int(video_data.loc[i, 'NA_Sales'])
            platform_sales[index]['EU_Sales'] += int(video_data.loc[i, 'EU_Sales'])
            platform_sales[index]['JP_Sales'] += int(video_data.loc[i, 'JP_Sales'])
            platform_sales[index]['Global_Sales'] += int(video_data.loc[i, 'Global_Sales'])
        else:
            obj = {
                'Platform': video_data.loc[i, 'Platform'],
                'NA_Sales': int(video_data.loc[i, 'NA_Sales']),
                'EU_Sales': int(video_data.loc[i, 'EU_Sales']),
                'JP_Sales': int(video_data.loc[i, 'JP_Sales']),
                'Global_Sales': int(video_data.loc[i, 'Global_Sales'])
            }
            platform_sales.append(obj)

platform_sales_df = pd.DataFrame(platform_sales)

platform_sales_df
```

```python
platform_year_sales = []

for i in video_data.index:
    if len(platform_year_sales) == 0:
        obj = {
            'Platform': video_data.loc[i, 'Platform'],
            'Year': int(video_data.loc[i, 'Year']),
            'NA_Sales': int(video_data.loc[i, 'NA_Sales']),
            'EU_Sales': int(video_data.loc[i, 'EU_Sales']),
            'JP_Sales': int(video_data.loc[i, 'JP_Sales']),
            'Global_Sales': int(video_data.loc[i, 'Global_Sales'])
        }
        platform_year_sales.append(obj)
    else:
        status = False
        for item in platform_year_sales:
            if item['Platform'] == video_data.loc[i, 'Platform'] and item['Year'] == video_data.loc[i, 'Year']:
                index_p = platform_year_sales.index(item)
                platform_year_sales[index_p]['NA_Sales'] += video_data.loc[i, 'NA_Sales']
                platform_year_sales[index_p]['EU_Sales'] += video_data.loc[i, 'EU_Sales']
                platform_year_sales[index_p]['JP_Sales'] += video_data.loc[i, 'JP_Sales']
                platform_year_sales[index_p]['Global_Sales'] += video_data.loc[i, 'Global_Sales']
                status = True
                break
        if not status:
            obj = {
                'Platform': video_data.loc[i, 'Platform'],
                'Year': int(video_data.loc[i, 'Year']),
                'NA_Sales': int(video_data.loc[i, 'NA_Sales']),
                'EU_Sales': int(video_data.loc[i, 'EU_Sales']),
                'JP_Sales': int(video_data.loc[i, 'JP_Sales']),
                'Global_Sales': int(video_data.loc[i, 'Global_Sales'])
            }
            platform_year_sales.append(obj)

platform_year_sales_df = pd.DataFrame(platform_year_sales)

platform_year_sales_df.sort_values('Year')
```
